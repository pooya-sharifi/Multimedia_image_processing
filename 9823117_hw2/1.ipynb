```python
import numpy as np
import cv2
import matplotlib.pyplot as pyplot
```

```python
# Read the image
image = cv2.imread('../HW2/AerialView.jpeg')

# cv2.imshow('Image with Lines', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
pyplot.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
pyplot.show()
```

```python
sv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

lower_green = np.array([30, 0, 0])
upper_green = np.array([90, 255, 255])
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
mask = cv2.inRange(hsv, lower_green, upper_green)
res = cv2.bitwise_and(image, image, mask= mask)

pyplot.imshow(mask)
pyplot.show()
```

```python
pyplot.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
pyplot.show()
```

```python
contours, hierarchy=cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(image,contours,-1,(0,0,255),5)
```

```python
pyplot.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
pyplot.show()
```
