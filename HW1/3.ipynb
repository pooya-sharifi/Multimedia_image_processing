```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
import cv2
import numpy as np

# Load image
img = cv2.imread('./img3.jpg')

imshow("original",img)
```

```python
# Create structuring element
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

# Erode image
eroded_img = cv2.erode(img, kernel, iterations=2)

imshow("erode_1",eroded_img)

# # Erode image
# eroded_img = cv2.erode(eroded_img, kernel, iterations=1)

# imshow("erode_2",eroded_img)
```

```python
dilated = cv2.dilate(eroded_img, kernel, iterations = 2)

imshow("dilated_1",dilated)
```

```python
# Perform closing to fill gaps and make dots rounder
closed = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel)

imshow("closed",closed)
```

```python
lines=img-dilated
imshow("lines",lines)
kernel_2 = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
eroded_img_2 = cv2.erode(lines, kernel_2, iterations=1)
dilated_2 = cv2.dilate(eroded_img_2, kernel_2, iterations = 1)
imshow("lines",dilated_2)
```
