```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
def imshow_2(title = "", image = None):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.show()
```

```python
img = cv2.imread('./img1.tif')

imshow("img",img)

normalized_img=img/255
print(normalized_img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray=gray/255
imshow_2("grey",gray)


print(gray)
# imshow("normalized",normalized_img)
```

```python
import numpy as np
```

```python
# Create zeros array to store new image
new_im_1 = np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image 
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        # print(img[i,j])
        if(gray[i,j]<0.2):
            new_im_1[i,j]=-255*gray[i,j]
        if(0.2<gray[i,j]<0.55):
            new_im_1[i,j]=-255*gray[i,j]
        if(0.55<gray[i,j]):
            new_im_1[i,j]=-255*gray[i,j]
```

```python
imshow_2("1 image",np.array(new_im_1)/255)
```

```python
# Create zeros array to store new image
new_im_2 = np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image 
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        # print(img[i,j])
        if(gray[i,j]<0.2):
            new_im_2[i,j]=255*gray[i,j]
        if(0.2<gray[i,j]<0.55):
            new_im_2[i,j]=255*0
        if(0.55<gray[i,j]):
            new_im_2[i,j]=255*gray[i,j]
        
        
```

```python
imshow_2("2 image",np.array(new_im_2)/255)
```

```python
# Create zeros array to store new image
new_im_3 = np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image 
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        # print(img[i,j])
        if(gray[i,j]<0.4):
            new_im_3[i,j]=0
        if(0.4<gray[i,j]<0.55):
            new_im_3[i,j]=255*gray[i,j]
        if(0.55<gray[i,j]):
            new_im_3[i,j]=255*1
```

```python
imshow_2("3 image",np.array(new_im_3)/255)
```

```python
# Create zeros array to store new image
new_im_4 = np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image 
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        # print(img[i,j])
        if(gray[i,j]<0.4):
            new_im_4[i,j]=255*3/2*gray[i,j]
        if(0.4<gray[i,j]<0.55):
            new_im_4[i,j]=255*0.3
        if(0.55<gray[i,j]):
            new_im_4[i,j]=255*(0.7/0.45*gray[i,j]-0.555)
```

```python
imshow_2("4 image",np.array(new_im_4)/255)
```

```python
# Create zeros array to store new image
new_im_5 = np.zeros((gray.shape[0],gray.shape[1]),dtype = 'uint8')
 
# Loop over the image 
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        # print(img[i,j])
        if(gray[i,j]<0.4):
            new_im_4[i,j]=255*gray[i,j]**1/2
        if(0.4<gray[i,j]<0.55):
            new_im_4[i,j]=255*gray[i,j]**1/2
        if(0.55<gray[i,j]):
            new_im_4[i,j]=255*gray[i,j]**1/2
```

```python
imshow_2("5 image",np.array(new_im_4)/255)
```
