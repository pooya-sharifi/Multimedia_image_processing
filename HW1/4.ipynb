```python
import cv2
from matplotlib import pyplot as plt
```

```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
import cv2
import numpy as np

# Load image
img = cv2.imread('./img4.jpg')

imshow("original",img)
```

```python
img_new=np.zeros((img.shape[0],img.shape[1]),dtype = 'uint8')
median_new = cv2.medianBlur(img, 5)
imshow("image with median blur",median_new)
```

```python
# dst = cv2.GaussianBlur(img, 5, borderType=BORDER_DEFAULT]]] )
ksize = (5, 5)
sigmaX = 1.5
guassian_img = cv2.GaussianBlur(img, ksize, sigmaX)
imshow("guassian blured image",guassian_img)
```

```python
bilateral_img = cv2.bilateralFilter(img,9,75,75)
imshow("bilateral filter",bilateral_img)
```

```python
# Setting parameter values
t_lower = 50  # Lower Threshold
t_upper = 150  # Upper threshold
  
# Applying the Canny Edge filter
line_of_og = cv2.Canny(img, t_lower, t_upper)

imshow("line of OG",line_of_og)
```

```python
# Setting parameter values
t_lower = 50  # Lower Threshold
t_upper = 150  # Upper threshold
  
# Applying the Canny Edge filter
line_of_median = cv2.Canny(median_new, t_lower, t_upper)

imshow("line_of_median",line_of_median)
```

```python
# Setting parameter values
t_lower = 50  # Lower Threshold
t_upper = 150  # Upper threshold
  
# Applying the Canny Edge filter
line_of_guas = cv2.Canny(guassian_img, t_lower, t_upper)

imshow("line_of_guas",line_of_guas)
```

```python
# Setting parameter values
t_lower = 50  # Lower Threshold
t_upper = 150  # Upper threshold
  
# Applying the Canny Edge filter
line_of_bi = cv2.Canny(bilateral_img, t_lower, t_upper)

imshow("line_of_bi",line_of_bi)
```
